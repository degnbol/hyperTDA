## Scale simulated and invivo tracks to have the average distance between adjacent points equal to one

```python
def md(c):
    """
    Computes average displacement of a curve c
    """
    l = [np.linalg.norm(c[i]-c[i+1]) for i in range(len(c)-1)]
    r = np.average(np.array(l))
    return r
```
